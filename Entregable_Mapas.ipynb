```python
import matplotlib
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv("avocado.csv")

data.head()
```

```python
data.describe()
```

```python
average_price = data['AveragePrice']
total_volume = data['Total Volume']
num_4046 = data['4046']
num_4225 = data['4225']
num_4770 = data['4770']
total_bags = data['Total Bags']
small_bags = data['Small Bags']
large_bags = data['Large Bags']
xlarge_bags = data['Large Bags']
year = data['year']

numeric_values = [average_price,total_volume,num_4046,num_4225,num_4770,total_bags,small_bags,large_bags,xlarge_bags,year]

fig = plt.figure(figsize =(10, 7)) # Crear dimensiones de la figura
plt.boxplot(numeric_values) # Plot los datos que queremos, columna específica

plt.title('Box and whiskers de valores numéricos de avocado.csv \n', 
          fontweight ="bold")

plt.xlabel('Eje X (Columnas)')
plt.ylabel('Eje Y / x10^7')

plt.show() # Mostrar la figura
```

```python
plt.hist(data['AveragePrice'],histtype ='bar', rwidth= 0.4, color='pink')
plt.title('Average Price \n', 
          fontweight ="bold")
plt.xlabel('Average price') 
plt.ylabel('Eje Y')
plt.show()
```

```python
plt.hist(data['Total Volume'],histtype ='bar', rwidth= 0.4, color='pink')
plt.title('Total Volume \n', 
          fontweight ="bold")
plt.xlabel('Total volume') 
plt.ylabel('Eje Y')
plt.show()
```

```python
bags = [total_bags,small_bags,large_bags,xlarge_bags]

label_bags = ['Total Bags','Small Bags','Large Bags','Xlarge Bags']

colors = ['pink','blue','purple','green']

plt.hist(bags, histtype ='bar', color = colors)

plt.title('Total Bags, Small Bags, Large Bags, Xlarge Bags\n', 
          fontweight ="bold") 

plt.xlabel('Total Bags (pink), Small Bags (blue), Large Bags (purple), Xlarge Bags (green)')
  
plt.show()
```
